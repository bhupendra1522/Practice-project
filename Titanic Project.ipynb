```python
df["Salutation"] = df["Name"].str.extract('([A-Za-z]+)\.', expand = False)
```

```python
df["Salutation"].unique() 
```

```python
df.loc[df["Age"].isnull()]["Salutation"].unique() 
```

```python
df['Age'] = df.groupby('Salutation')['Age'].apply(lambda x:x.fillna(x.median()))
```

```python
df.info()                    
```

```python
from sklearn.ensemble import RandomForestClassifier
y = df["Survived"]
features = ["Sex", "Pclass", "SibSp", "Parch", "Embarked", "Age"]
x = pd.get_dummies(df[features])
x_test = pd.get_dummies(df[features])
model = (RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1))
model.fit(x, y)
model.score(x, y)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV

y = df["Survived"]
features = ["Sex", "Pclass", "SibSp", "Parch", "Embarked", "Age"]
X = pd.get_dummies(df[features])
model = RandomForestClassifier(random_state=1)
param_grid = {
    'n_estimators': [50, 75, 100, 200],
    'max_depth': [4, 5, 6, 8]
}
model_CV = GridSearchCV(estimator=model, param_grid=param_grid, cv=5)
model_CV.fit(X, y)
print(model_CV.score(X, y))
model_CV.best_params_
```
